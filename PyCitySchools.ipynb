# PyCity Schools Analysis

 - The school district has 15 high schools; 7 district schools and 8 charter schools.
 - Each of the district schools has over 2,700 students while each of the charter schools has
     less than 2,300 students.
 - Overall passing rates for the district schools were less than 75% per school. Overall passing
     rates for the charter schools were greater than 94% per school.
 - In particular, average math scores were lower at the district schools (76.6%-77.3%) than at
     the charter schools (> 83%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Files to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

# School District Summary

```python
# Select the appropriate columns
school_data_complete=school_data_complete[['grade','school_name','reading_score','math_score','type','size','budget','Student ID']]
# school_data_complete.head()
```

```python
# Select the unique school names
school_names = school_data_complete['school_name'].unique()
# school_names
```

```python
# Count the number of schools
count_schools = len(school_names)
count_schools
```

```python
# Count the total number of students
tot_students = school_data_complete['Student ID'].count()
tot_students
```

```python
# Get the total budget for all schools
tot_budget = school_data_complete['budget'].unique().sum()
tot_budget
```

```python
# Get the average math score for all students
avg_math = school_data_complete['math_score'].sum() / tot_students
avg_math
```

```python
# Get the average reading score for all students
avg_reading = school_data_complete['reading_score'].sum() / tot_students
avg_reading
```

```python
# Total number of students passing math
ps_math = school_data_complete['math_score'] >= 70
#Total number of students passing math
pass_math = np.sum(ps_math)
pass_math
```

```python
# Total number of students passing reading
ps_read = school_data_complete['reading_score'] >= 70
#Total number of students passing reading
pass_reading = np.sum(ps_read)
pass_reading
```

```python
#Percent of students passing math
percent_math = pass_math/tot_students*100
percent_math
```

```python
#Percent of students passing reading
percent_read = pass_reading/tot_students*100
percent_read
```

```python
#Average percent of students passing math and reading
tot_rate = (percent_math + percent_read) / 2
tot_rate
```

```python
# Create a dataframe with the above stats
dist_smry = {"Total Schools":[count_schools],
            "Total Students":[tot_students],
            "Total Budget":[tot_budget],
            "Average Math Score":[avg_math],
            "Average Reading Score":[avg_reading],
            "% Passing Math":[percent_math],
            "% Passing Reading":[percent_read],
            "% Overall Passing Rate":[tot_rate]
               }
dist_smry_df = pd.DataFrame(dist_smry)
dist_smry_df["Total Students"] = dist_smry_df["Total Students"].map("${:,}".format)
dist_smry_df["Total Budget"] = dist_smry_df["Total Budget"].map("${:,}".format)
dist_smry_df["Average Math Score"] = dist_smry_df["Average Math Score"].map("{:.1f}".format)
dist_smry_df["Average Reading Score"] = dist_smry_df["Average Reading Score"].map("{:.1f}".format)
dist_smry_df["% Passing Math"] = dist_smry_df["% Passing Math"].map("{:.1f}".format)
dist_smry_df["% Passing Reading"] = dist_smry_df["% Passing Reading"].map("{:.1f}".format)
dist_smry_df["% Overall Passing Rate"] = dist_smry_df["% Overall Passing Rate"].map("{:.1f}".format)
dist_smry_df
```

# School Summaries

```python
# Determine the school type
sch_type = school_data.set_index(["school_name"])["type"]
sch_type
```

```python
# Count the number of students in each school
count_per_school = school_data_complete['school_name'].value_counts()
count_per_school
```

```python
# Average budget per school
bdgt_per_school = school_data_complete.groupby(['school_name']).mean()['budget']
bdgt_student_per_school = bdgt_per_school / count_per_school
bdgt_student_per_school
```

```python
# Average math score per school
avg_mscore = school_data_complete.groupby(["school_name"]).mean()["math_score"]
avg_mscore
```

```python
# Average reading score per school
avg_rscore = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
avg_rscore
```

```python
# Passing math by school
math_passing = school_data_complete[(school_data_complete['math_score'] >= 70)]
# math_passing.head()
```

```python
# Passing reading by school
reading_passing = school_data_complete[(school_data_complete['reading_score'] >= 70)]
# reading_passing.head()
```

```python
# Percent passing math or reading by school
math_pass_per_school = math_passing.groupby(['school_name']).count()['Student ID'] / count_per_school*100
# math_pass_per_school
reading_pass_per_school = reading_passing.groupby(['school_name']).count()['Student ID'] / count_per_school*100
# reading_pass_per_school
```

```python
# Overall percent passing by school
overall_passing = (math_pass_per_school + reading_pass_per_school) / 2.0
# overall_passing
```

```python
# Create dataframe
school_smry_df = pd.DataFrame({"School Type": sch_type,
               "Total Students": count_per_school,
               "Total School Budget": bdgt_per_school,
               "Per Student Budget": bdgt_student_per_school,
               "Average Math Score": avg_mscore,
               "Average Reading Score": avg_rscore,
               "% Passing Math": math_pass_per_school,
               "% Passing Reading": reading_pass_per_school,
               "% Overall Passing Rate": overall_passing
               })
# Format dataframe
school_smry_df["Total School Budget"] = school_smry_df["Total School Budget"].map("${:,.0f}".format)
school_smry_df["Total Students"] = school_smry_df["Total Students"].map("{:,}".format)
school_smry_df["Per Student Budget"] = school_smry_df["Per Student Budget"].map("${:.0f}".format)
school_smry_df["Average Math Score"] = school_smry_df["Average Math Score"].map("{:,.1f}".format)
school_smry_df["Average Reading Score"] = school_smry_df["Average Reading Score"].map("{:,.1f}".format)
school_smry_df["% Passing Math"] = school_smry_df["% Passing Math"].map("{:,.1f}".format)
school_smry_df["% Passing Reading"] = school_smry_df["% Passing Reading"].map("{:,.1f}".format)
school_smry_df["% Overall Passing Rate"] = school_smry_df["% Overall Passing Rate"].map("{:,.1f}".format)
school_smry_df
```

# Student Budget by School

```python
# Sort data by per student budget amounts
cost_per_student = school_smry_df.sort_values('Per Student Budget')
cost_per_student
```

# Passing Rate by School

```python
# Sort data by overall passing rates
passing_rate = school_smry_df.sort_values('% Overall Passing Rate')
passing_rate
```

# Math Scores by Grade

```python
# Aggregate grade levels
grade9 = school_data_complete[(school_data_complete["grade"] == "9th")]
grade10 = school_data_complete[(school_data_complete["grade"] == "10th")]
grade11 = school_data_complete[(school_data_complete["grade"] == "11th")]
grade12 = school_data_complete[(school_data_complete["grade"] == "12th")]
```

```python
# Average math scores by school and grade
mscores_grade9 = grade9.groupby(["school_name"]).mean()["math_score"]
mscores_grade10 = grade10.groupby(["school_name"]).mean()["math_score"]
mscores_grade11 = grade11.groupby(["school_name"]).mean()["math_score"]
mscores_grade12 = grade12.groupby(["school_name"]).mean()["math_score"]
# mscores_grade12
```

```python
# Create a dataframe for the average math scores
mscores_df = pd.DataFrame({"Ninth Grade": mscores_grade9, "Tenth Grade": mscores_grade10,
                           "Eleventh Grade": mscores_grade11, "Twelfth Grade": mscores_grade12})
# Format dataframe
mscores_df["Ninth Grade"] = mscores_df["Ninth Grade"].map("{:,.1f}".format)
mscores_df["Tenth Grade"] = mscores_df["Tenth Grade"].map("{:,.1f}".format)
mscores_df["Eleventh Grade"] = mscores_df["Eleventh Grade"].map("{:,.1f}".format)
mscores_df["Twelfth Grade"] = mscores_df["Twelfth Grade"].map("{:,.1f}".format)
mscores_df.index.name = None
mscores_df
```
